# CSV (Conceptual)

```python
data=('col1,col2,col3\n'
     'x,y,1\n'
     'a,b,2\n'
     'c,d,3\n')
```

```python
from io import StringIO,BytesIO
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))
```

```python
## Readd from specific column
df2=pd.read_csv(StringIO(data),usecols=['col1','col3'])
```

```python
df2
```

```python
df2.to_csv('Test1.csv')
```

```python
# Specifying Column Data Types
data2=('a,b,c,d\n'
       '1,2,3,4\n'
       '5,6,7,8\n'
       '9,10,11,12'
)
```

```python
print(data2)
```

```python
df4=pd.read_csv(StringIO(data2),dtype=int)
```

```python
df4
```

```python
df4['a'][1]
```

```python
df4['a']
```

```python
df4=pd.read_csv(StringIO(data2),dtype={'b':int,'c':float,'a':'Int64'})
```

```python
df4
```

```python
df4.dtypes
```

```python
# Index Columns
data5=('Index,a,b,c\n'
       '4,apple,bat,5.7\n'
       '8,orange,cow,10'
)
```

```python
pd.read_csv(StringIO(data5),index_col=0)
```

```python
data6=('a,b,c\n'
       '4,apple,bat\n'
       '8,orange,cow'
)
```

```python
pd.read_csv(StringIO(data6),index_col=False)
```

```python
## Combining usecols & index_col
data6=('Index,a,b,c\n'
       '4,apple,bat,5.7\n'
       '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data6),usecols=['b','c'],index_col=False)
```

```python
## Quoting & Escaping Characters . Very Helpful in NLP
data7='a,b\n"hello,\\"Bob\\",nice to see you",5'
```

```python
pd.read_csv(StringIO(data7),escapechar="\\")
```
