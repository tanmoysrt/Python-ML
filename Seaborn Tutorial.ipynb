```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head() # tip is dependent feature
```

```python
df.dtypes
```

## Correlation with Heatmap

### A coorelation heatmap uses coloured cells , typically in a monochromatic scale, to show a 2D correolation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection 

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Joinplot

### Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['sex'].value_counts()
```

## Dist Plot
##### Dist plot helps us to check the distribution of columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Categorical Plots

##### Seaborn also helps us in doing  the analysis on Catehorical Data Points .
##### 1. boxplot
##### 2. viollnplot
##### 3.countplot
##### 4.bar plot


```python
## Count Plot
sns.countplot('sex',data=df)
```

```python
## Count Plot
sns.countplot(y='sex',data=df)
```

```python
## Count Plot
sns.countplot('day',data=df)
```

```python
## Bar Plot
sns.barplot(x='total_bill',y='sex',data=df)
```

## Box Plot

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df)
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

## Violin Plot

```python
sns.violinplot(x='total_bill',y='day',data=df)
```
